```python
from lambeq import SpacyTokeniser, BobcatParser, SpiderAnsatz, AtomicType, IQPAnsatz
from lambeq.backend.tensor import Dim
```

```python
tokeniser = SpacyTokeniser()
sentence = "A warm welcome from Vijay and team!"
token = tokeniser.tokenise_sentence(sentence)
token
```

```python
parser = BobcatParser(verbose="suppress")
diagram = parser.sentence2diagram(token, tokenised='true')
diagram.draw(figsize=(23,4), fontsize=12)
```

```python
ansatz = SpiderAnsatz({AtomicType.NOUN: Dim(2), AtomicType.SENTENCE: Dim(2)})

train_circuits = [ansatz(diagram)]
dev_circuits = [ansatz(diagram)]
test_circuits = [ansatz(diagram)]

train_circuits[-1].draw(figsize=(23, 4))
```

```python
ansatz = IQPAnsatz({AtomicType.NOUN: 1, AtomicType.SENTENCE: 1}, n_layers=2)
circuit = ansatz(diagram)
circuit.draw(figsize=(23, 15))
```
